```python
#loading the library
import pandas as pd

df = pd.read_excel("../data/final_pricing_dataset.xlsx",header =0)

```

```python
df = df.drop("month", axis=1)

df["demand"] = df["demand"] + (df["temperature"] - 25) * 2
df["demand"] = df["demand"].clip(lower=0)
```

```python
####### DATA CLEANING #######
df["date"] = pd.to_datetime(df["date"]).dt.date
```

```python
df["profit"] = (df["price"] - df["cost"]) * df["demand"]
```

```python
import plotly.express as px
```

```python
# DEMAND DISTRIBUTION

fig = px.histogram(df[df["demand"] < 100],
                x="demand",
                nbins=50,
                title="Demand Distribution (Filtered)",
                color_discrete_sequence=["#636EFA"])

fig.show()
```

```python
# PRICE VS DEMAND 

fig = px.scatter((df[(df["price"]< 50) & (df["demand"] < 100)]),
                x="price",
                y="demand",
                color="demand",
                title="Price vs Demand",
                color_continuous_scale="viridis")

fig.show()
```

```python
# PROMO VS DEMAND
promo_df = df.groupby("promo")["demand"].mean().reset_index()

fig = px.bar(promo_df,
            x="promo",
            y="demand",
            color="promo",
            title="Promo vs Demand",
            color_discrete_sequence=["#EF553B", "#00CC96"])

fig.show()
```

```python
#PROFIT DISTRIBUTION

fig = px.histogram(df[df["profit"]< 30],
                x="profit",
                nbins=50,
                title="Profit Distribution",
                color_discrete_sequence=["#AB63FA"])

fig.show()
```

```python
# TEMPERATURE VS DEMAND

fig = px.scatter(df,
                x="temperature",
                y="demand",
                color="temperature",
                title="Temperature vs Demand",
                color_continuous_scale="plasma")

fig.show()
```

```python
#WEEKEND VS DEMAND
week_df = df.groupby("is_weekend")["demand"].mean().reset_index()

fig = px.bar(week_df,
            x="is_weekend",
            y="demand",
            color="is_weekend",
            title="Weekend vs Demand",
            color_discrete_sequence=["#FFA15A", "#19D3F3"])

fig.show()
```

```python
#correlation heatmap

import plotly.figure_factory as ff

corr = df.select_dtypes(include="number").corr()

fig = px.imshow(
    corr,
    text_auto=True,
    aspect="auto",
    color_continuous_scale="viridis",
    title="Correlation Heatmap"
)

fig.update_layout(
    width=800,
    height=700
)

fig.show()
```
